```python
url = "https://www.naver.com/"
txt = requests.get(url) # 웹페이지 그냥 긁자로 읽어오기
html = bs(txt.text) # 읽어온 글자를 의미가 있는 bs로 html 변환
```

```python
## 벅스 차트 읽어오기 

url = "https://music.bugs.co.kr/chart"
txt = requests.get(url)
html = bs(txt.text)

```

```python
len(html.select('tr'))
```

```python
len(html.select('table.byChart > tbody > tr')) # .은 클래스를 의미
# 최종 원하는 100곡
```

```python
# 전체 소스중 위에서 찾아낸 100개의 tr 들만 추출
songs = html.select('table.byChart > tbody > tr')
print(songs)
```

```python
song = songs[0]
print(song)
```

```python
len(song.select('a')) 
```

```python
len(song.select('p > a'))
```

```python
song.select('p.title > a')
```

```python
print(song.select('p.title > a')[0].text)
```

```python
# 1등 노래제목 추출 완료
# 2등 노래제목 추출하기

print(songs[1].select('p.title > a')[0].text)
```

```python
# 1등의 가수명
print(songs[0].select('p.artist > a')[0].text)
```

```python
# 1등부터 100등까지 노래제목과 가수명 추출
song_data = []
rank = 1

for song in songs:
    
    title = song.select('p.title > a')[0].text
    singer = song.select('p.artist > a')[0].text

    song_data.append(['Bugs', rank, title, singer])
    rank += 1 

import pandas as pd
df = pd.DataFrame(song_data, columns = ['서비스', '순위', '타이틀', '가수'])
df


    
```

```python
# df을 excel 파일로 변환

df.to_excel('bugs.xlsx', index=False)
```
