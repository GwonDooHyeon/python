```python
# 전체 소스중 위에서 찾아낸 100개의 tr 들만 추출
songs = html.select('table.byChart > tbody > tr')
print(songs)
```

```python
# 1등 노래제목 추출 완료
# 2등 노래제목 추출하기

print(songs[1].select('p.title > a')[0].text)
```

```python
# 1등의 가수명
print(songs[0].select('p.artist > a')[0].text)
```

```python
# 1등부터 100등까지 노래제목과 가수명 추출
song_data = []
rank = 1

for song in songs:
    
    title = song.select('p.title > a')[0].text
    singer = song.select('p.artist > a')[0].text

    song_data.append(['Bugs', rank, title, singer])
    rank += 1 

import pandas as pd
df = pd.DataFrame(song_data, columns = ['서비스', '순위', '타이틀', '가수'])
df


    
```

```python
# df을 excel 파일로 변환
df.to_excel('bugs.xlsx',index=False)
```

```python
from bs4 import BeautifulSoup as bs # 읽어온 웹페이지 정제
import pandas as pd                 # 데이터 분석 모듈 필수!!
from selenium import webdriver  
```

```python
driver = webdriver.Chrome('chromedriver.exe')  #크롬 드라이버 실행
driver.get("https://www.genie.co.kr/chart/top200?ditc=D&ymd=20230227&hh=16&rtm=Y&pg=1") #드라이버로 크롤링할 페이지

txt = driver.page_source  #드라이버로 얻어온 페이지를 txt로
html = bs(txt) # 벅스 문법과 약간 다름


songs = html.select('tbody > tr') 
song_data = [] #song_data라는 빈 배열을 생성
rank = 1

for song in songs:
    
    title = song.select('a.title.ellipsis')[0].text.lstrip()
    singer = song.select('a.artist.ellipsis')[0].text

    song_data.append(['Genie', rank, title, singer])
    rank += 1  


    
# 1페이지 50위까지 반복해서 배열 song_data에 추가

driver = webdriver.Chrome('chromedriver.exe')
driver.get("https://www.genie.co.kr/chart/top200?ditc=D&ymd=20230227&hh=16&rtm=Y&pg=2")

txt = driver.page_source
html = bs(txt) # 벅스 문법과 약간 다름

songs = html.select('tbody > tr')

for song in songs:
    
    title = song.select('a.title.ellipsis')[0].text.lstrip()
    singer = song.select('a.artist.ellipsis')[0].text

    song_data.append(['Genie', rank, title, singer])
    rank += 1
# 2페이지를 똑같이 추가해서 배열에 추가    

df = pd.DataFrame(song_data, columns = ['서비스','순위','타이틀','가수']) #추가한 배열을 df에 넣으면 된다
df
```
